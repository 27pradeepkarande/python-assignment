```python
3.Create a python program to compute the electricity bill for a household.
ans :
    def calculate_electricity_bill(units):
    bill = 0
    rates = [(100, 4.5), (100, 6), (100, 10), (float('inf'), 20)]
    remaining_units = units

    for limit, rate in rates:
        if remaining_units > 0 :
            units_to_charge = min(remaining_units, limit)
            bill += units_to_charge * rate
            remaining_units -= units_to_charge
        else:
            break

    return bill

# Take the units of electricity consumed in a month from the user as input
units_consumed = float(input("Enter the number of units consumed: "))

# Calculate the bill
bill_amount = calculate_electricity_bill(units_consumed)

# Print the bill amount
print(f"The electricity bill for {units_consumed} units is: Rs. {bill_amount:.2f}")
```

```python
4. Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate
the cube of each number and if the cube of that number is divisible by 4 or 5 then append
that number in a list and print that list.
ans : 
    #by using for loop
    
    # Create a list of numbers from 1 to 100
    numbers = list(range(1, 101))

   # Initialize an empty list to store the numbers whose cubes are divisible by 4 or 5
   result_for = []

  # Iterate through the list of numbers
  for num in numbers:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        result_for.append(num)

  # Print the resulting list
  print("Using for loop:", result_for)

  
  # by using while loop
    
   # Create a list of numbers from 1 to 100
   numbers = list(range(1, 101))

   # Initialize an empty list to store the numbers whose cubes are divisible by 4 or 5
   result_while = []

   # Initialize the starting index
   i = 0

   # Iterate through the list using a while loop
    while i < len(numbers):
        num = numbers[i]
        cube = num ** 3
        if cube % 4 == 0 or cube % 5 == 0:
        result_while.append(num)
    i += 1

# Print the resulting list
print("Using while loop:", result_while)
```

```python
Q5. Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"

ans :
    # Given string
string = "I want to become a data scientist"

# List of vowels
vowels = "aeiouAEIOU"

# Initialize a counter for vowels
vowel_count = 0

# Iterate through each character in the string
for char in string:
    if char in vowels:
        vowel_count += 1

# Print the total count of vowels
print(f"The number of vowels in the given string is: {vowel_count}")
```
